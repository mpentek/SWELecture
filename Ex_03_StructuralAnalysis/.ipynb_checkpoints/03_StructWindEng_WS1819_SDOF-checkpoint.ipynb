#### Creating the time instances as an array

The start time, end time and the number of time steps are specified here for generating the time series.

### 3.1 Dynamic Analysis

The response of MDOF under dynamic loading is computed by different time integration. Three time integration is presented in this section. 
1. __Generalised-Alpha__
2. __Euler First Order__
2. __Euler First and Second Order__

_THE OBJECT-ORIENTED GENERALIZED-ALPHA SOLVER
Implementation adapted from I. Hanzlicek (2014). Original implementation by M. Andre described in: Formulation of the Generalized-Alpha method for LAGRANGE. Technical Report, Chair of Structural Analysis @TUM, 2012.
See J. Chung, G.M. Hulbert: A time integration algorithm for structural dynamics
wih improved numerical dissipation: the generalized-aplha mehod. ASME J. Appl. 
Mech., 60:371-375,1993._

_THE EULER ALGORITHM USING FIRST AND SECOND ORDER APPROXIMATION
Implementation of the well-known finite difference approach, theory also
described in J.L. Cieslinski, B. Ratkiewicz: On the Simulations of the Classical
Harmonic Oscillator Equations by Difference Equations, PY 502, Hindawi Publishing 
Corporation, Advances in Difference Equations, Volume 2006. An algorithmic description
can also be found in H.P. Gavin: Numerical Integration in Structural Dynamics,
CEE 541, Structural Dynamics, Department of Civil & Environmental Engineering, 
Duke University Fall 2016._

###### Structural setup 

```python
# mass
M = 0.1
# eigenfrequency
eigenFreq = 10.0
# stiffness
K = M * (eigenFreq * 2 * np.pi)**2
xi = 0.00 # damping ratio
B = xi * 2.0 * np.sqrt(M * K)
# zero damping in this case
```

######  Initial conditions

```python
u0 = 1.0
v0 = 0.0
a0 = 0.0
```

###### External loading  

Three types of loads are defined here. 
1. Free vibration case - No external loads
2. Harmonic excitation
3. superimposed signal as in previous example. 

```python
# sine with given amplitude = 1 and frequency = 10 Hz
freq = 10
amplSin = 1
signalSin = amplSin * np.sin( 2*np.pi*freq * arrayTime)

# normal random signal with given mean m = 0 and standard dev sd = 0.25 ->
signalRand = np.random.normal(0, 0.25, len(arrayTime))

# constant signal with given amplitude = 10
amplConst = 10
signalConst = amplConst * np.ones(len(arrayTime))

# superposing the above signals 
# superposition weighting
coefSignal1 = 1
coefSignal2 = 0.25
coefSignal3 = 0.25
signalSuperposed = coefSignal1 * signalConst + coefSignal2 * signalSin + coefSignal3 * signalRand

signalNone = np.zeros(len(arrayTime))

# the external load 
extForce = signalNone
```

###### Let's plot the excitation function 

```python
#plot for force - for top dof
plt.figure(figsize=(15, 4))
plt.plot(arrayTime, extForce, "-k", lw=0.5)
plt.ylabel('Force [N]')
plt.xlabel('Time [s]')
plt.title("Force - external")
plt.grid(True)
```

###### Analytical solutions  

Analytical solutions are available for 
1. Undamped free vibration with initial displacement 
2. Undamped force vibration under harmonic excitation with no initial displacement 

Please refer A.K. Chopra: Dynamics of Structures: Theory and Applications to Earthquake Engineering, Prentice-Hall, 1995 for detailed descriptions .

```python
## undamped harmonic oscillation with only initial displacement
omega = np.sqrt( K / M)
u0an = 1.0
A = u0an**2 + (v0/omega)**2
theta = 0 #numpy.atan(u0an*omega/v0)
dispAnalyticWoExtF = A* np.sin(np.multiply(omega,arrayTime) + theta)
dispAnalyticWoExtFLimit = u0an * np.ones(len(arrayTime))

## undamped harmonic oscillation with only external force
u0an = 0
# u0 = 0, v0 = 0, omegaF = omega
dispAnalyticWExtF = np.zeros(len(arrayTime))
dispAnalyticWExtFLimit = np.zeros(len(arrayTime))
for i in range(len(arrayTime)):
    dispAnalyticWExtF[i] = amplSin /2/K * np.sin(omega * arrayTime[i]) - amplSin /2/K * (omega * arrayTime[i]) * np.sin(omega * arrayTime[i])
    dispAnalyticWExtFLimit[i] = amplSin /2/K * (omega * arrayTime[i])
    
dOmegaRatio = 0.01
forcingFreq = np.arange(0.0, 3.0 + dOmegaRatio, dOmegaRatio)

def DetermineTransmissibilityFuncion(forcingFrequencyRatio, dampingRatio):
    transmissibilityFunction = np.zeros(len(forcingFrequencyRatio))
    for i in range(len(forcingFreq)):
        transmissibilityFunction[i] = np.sqrt(1 + (2 * dampingRatio * forcingFrequencyRatio[i])**2) / np.sqrt((1 - forcingFrequencyRatio[i]**2)**2 + (2 * dampingRatio * forcingFrequencyRatio[i])**2)    
    return transmissibilityFunction
```

###### Let's plot the function

For 1. Undamped free vibration with initial displacement

```python
xAxisEnd = endTime
plt.figure(figsize=(15, 4))
plt.title('Undamped free vibration with initial displacement u0 = ' + str(u0))
plt.plot(arrayTime, dispAnalyticWoExtF, "-k", lw=1)
plt.plot(arrayTime, dispAnalyticWoExtFLimit, "-.r", lw=1)
plt.plot(arrayTime, -dispAnalyticWoExtFLimit, "-.r", lw=1)
#plt.xlim([0, xAxisEnd])
plt.ylabel('Displacement [m]')
plt.xlabel('Time [s]')
plt.grid(True)
```

Observe the limiting values of the function

For 2. Undamped force vibration under harmonic excitation with no initial displacement

```python
plt.figure(figsize=(15, 4))
plt.title('Harmonic (sinusoidal) excitation of the undamped system without initial displacement')
plt.plot(arrayTime, dispAnalyticWExtF, "-k", lw=0.5)
plt.plot(arrayTime, dispAnalyticWExtFLimit, "-.r", lw=0.5)
plt.plot(arrayTime, -dispAnalyticWExtFLimit, "-.r", lw=0.5)
#plt.xlim([0, xAxisEnd])
plt.ylabel('Displacement [m]')
plt.xlabel('Time [s]')
plt.grid()
```

The dynamic amplification is dependent on the ratio of forcing frequency to eigen frequency

```python
plt.figure(figsize=(8, 6))
plt.title('Frequency response -> transmissibility')
plt.plot(forcingFreq, DetermineTransmissibilityFuncion(forcingFreq, 0.00001), "-r", lw=0.5)
plt.plot(forcingFreq, DetermineTransmissibilityFuncion(forcingFreq, 0.01), "-.k", label="xi = 0.01", lw=0.5)
plt.plot(forcingFreq, DetermineTransmissibilityFuncion(forcingFreq, 0.05), "--b", label="xi = 0.05", lw=0.5)
plt.plot(forcingFreq, DetermineTransmissibilityFuncion(forcingFreq, 0.1), ":g", label="xi = 0.1", lw=0.5)
plt.xlim([forcingFreq[0], forcingFreq[-1]])
plt.ylim([0.0, 60.0])
plt.ylabel('Transmissibility [-]')
plt.xlabel('Ratio of forcing frequency-to-eigenfrequency [-]')
plt.legend(loc="best")
plt.grid(True)
```

##### Time integration schemes 

Solving the equation of motion at each time step  

```python
pInf = 0.15 # numerical parameter -> only needed for the GeneralizedAlpha solver
# create an object: structure - to be used by the GeneralizedAlpha scheme
structure = StructureSDOF(dt, M, B, K, pInf, u0, v0, a0)
#structure.printSetup()
```

##### Tip: Have a look at "StructureSDoF.py" for details 

```python
## data for storing results
# using objects - dictionaries would also be a good option
# create a samplaData class
class sampleData(): pass

# initiate objects and labels
dataEuler1 = sampleData()
dataEuler1.label = "Euler 1st" 

dataEuler12 = sampleData()
dataEuler12.label = "Euler 1st & 2nd" 

dataGenAlpha = sampleData()
dataGenAlpha.label = "Gen Alpha" 

# lists to store the results
dataEuler1.disp = []
dataEuler1.acc  = []
dataEuler1.vel  = []

dataEuler12.disp = []
dataEuler12.acc  = []
dataEuler12.vel  = []

dataGenAlpha.disp = []
dataGenAlpha.acc  = []
dataGenAlpha.vel  = []
# ===============================================================================
# computation time for each method
# computation time for each method
dataEuler1.computTime = 0.0
dataEuler12.computTime = 0.0
dataGenAlpha.computTime = 0.0
    
# initial values
dataEuler1.disp.append(u0)
dataEuler1.vel.append(v0)
dataEuler1.acc.append(a0)

dataEuler12.disp.append(u0)
dataEuler12.vel.append(v0)
dataEuler12.acc.append(a0)

dataGenAlpha.disp.append(u0)
dataGenAlpha.vel.append(v0)
dataGenAlpha.acc.append(a0)

# more initial values for the time integration schemes
dataEuler1.un1 = u0
dataEuler1.vn1 = v0   
dataEuler1.an1 = a0  

dataEuler12.un2 = u0
dataEuler12.un1 = u0 - dt * v0 + dt**2 /2 * a0
dataEuler12.vn1 = v0
dataEuler12.an1 = a0 
```

###### Time loop : Computing the response at each time instant

```python
for i in range(1,len(arrayTime)):
    currentTime = arrayTime[i]
    
    #===========================================================================
    ## Euler 1st order   
    t = time.time()    

    # solve the time integration step 
    # first order approximation of acceleration and velocity, respectively    
    dataEuler1.un0 = dataEuler1.un1 + dt * dataEuler1.vn1
    dataEuler1.vn0 = dataEuler1.vn1 + dt * dataEuler1.an1
    dataEuler1.an0 = 1/M * (extForce[i] - B * dataEuler1.vn0 - K * dataEuler1.un0)

    # append results to list    
    dataEuler1.disp.append(dataEuler1.un0)
    dataEuler1.vel.append(dataEuler1.vn0)
    dataEuler1.acc.append(dataEuler1.an0)

    # update results    
    dataEuler1.un1 = dataEuler1.un0
    dataEuler1.vn1 = dataEuler1.vn0
    dataEuler1.an1 = dataEuler1.an0
    
    # elapsed time accumulated
    dataEuler1.computTime += time.time() - t
    
    #===========================================================================
    ## Euler 1st and 2nd order   
    t = time.time()    
 
    # solve the time integration step   
    # second order approximation of acceleration, first order for velocity
    #version 1 - eq. 5.3 
    #LHS = M
    #RHS = extForce[i-1] * dt**2 + dataEuler12.un1* (2*M - B * dt - K *dt**2) + dataEuler12.un2 * (-M + B * dt)

    #version 2 - eq. 5.4 from J.L. Cieslinski, B. Ratkiewicz or eq. 6 from H.P. Gavin 
    LHS = M + B * dt/2
    RHS = extForce[i-1] * dt**2 + dataEuler12.un1* (2*M - K *dt**2) + dataEuler12.un2 * (-M + B * dt /2)

    #version 3 - eq. 5.5
    #LHS = M + B * dt
    #RHS = extForce[i-1] * dt**2 + dataEuler12.un1* (2*M + B * dt - K *dt**2) + dataEuler12.un2 * (-M)

    dataEuler12.un0 = RHS/LHS
    
    dataEuler12.vn0 = (dataEuler12.un0 - dataEuler12.un2) /2 /dt
    dataEuler12.an0 = (dataEuler12.un0 - 2*dataEuler12.un1 + dataEuler12.un2) / dt**2

    # append results to list   
    dataEuler12.disp.append(dataEuler12.un0)
    dataEuler12.vel.append(dataEuler12.vn0)
    dataEuler12.acc.append(dataEuler12.an0)

    # update results
    dataEuler12.un2 = dataEuler12.un1    
    dataEuler12.un1 = dataEuler12.un0
    
    # elapsed time accumulated
    dataEuler12.computTime += time.time() - t   
    
    #===========================================================================
    ## Generalized Alpha     
    t = time.time()    

    # solve the time integration step
    structure.solveStructure(extForce[i])

    # append results to list
    dataGenAlpha.disp.append(structure.getDisplacement())
    dataGenAlpha.vel.append(structure.getVelocity())
    dataGenAlpha.acc.append(structure.getAcceleration())

    # update results
    structure.updateStructureTimeStep()
    
    # elapsed time accumulated
    dataGenAlpha.computTime += time.time() - t     
```

```python
## Print computation time
print('## Computation time ##')
print('Euler 1st order: ' + str(dataEuler1.computTime) + ' s')
print('Euler 1st & 2nd order: ' + str(dataEuler12.computTime) + ' s')
print('Generalized Alpha: ' + str(dataGenAlpha.computTime) + ' s') 
print()
 
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(3, 1, 1)
# subplot for displacement
plt.plot(arrayTime, dataEuler1.disp, "-k", label=dataEuler1.label, lw=0.5)
plt.plot(arrayTime, dataEuler12.disp, "-.r", label=dataEuler12.label, lw=0.5)
plt.plot(arrayTime, dataGenAlpha.disp, "--g", label=dataGenAlpha.label, lw=0.5)
plt.xlim([0, xAxisEnd])
plt.ylabel('Displacement [m]')
plt.axhline(y=u0, linewidth=2, color = 'r')
plt.axhline(y=-u0, linewidth=2, color = 'r')
plt.legend(loc="best")
grid()

# subplot for velocity

plt.subplot(3, 1, 2)
plt.plot(arrayTime, dataEuler1.vel, "-k", label=dataEuler1.label, lw=0.5)
plt.plot(arrayTime, dataEuler12.vel, "-.r", label=dataEuler12.label, lw=0.5)
plt.plot(arrayTime, dataGenAlpha.vel, "--g", label=dataGenAlpha.label, lw=0.5)
plt.xlim([0, xAxisEnd])
plt.ylabel('Velocity [m/s]')
grid()

# subplot for acceleration
plt.subplot(3, 1, 3)
plt.plot(arrayTime, dataEuler1.acc, "-k", label=dataEuler1.label, lw=0.5)
plt.plot(arrayTime, dataEuler12.acc, "-.r", label=dataEuler12.label, lw=0.5)
plt.plot(arrayTime, dataGenAlpha.acc, "--g", label=dataGenAlpha.label, lw=0.5)
plt.xlim([0, xAxisEnd])
plt.ylabel('Acceleration [m/s^2]')
grid()
```
