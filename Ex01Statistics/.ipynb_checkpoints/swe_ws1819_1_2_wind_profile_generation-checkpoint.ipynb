 Wind profile for the building at Terrain category I and IV


```python
# import
import matplotlib.pyplot as plt
import numpy as np
```

#### Gust wind speed computed from previous example for a return period of 50 years was 40.12 m/s^2. The mean wind speed is computed as

```python
gust_windspeed = 40.12
# 1.4 is the code specified factor to convert from gust to mean wind speed 
mean_windspeed = gust_windspeed/1.4 
```

Two terrain categories are assumed in this exercise.
1. The building is located at the seaside.__(Terrain category I)__
2. The building is located at an urban area with height of adjacent building upto 15m. __(Terrain category IV)__

Let us calculate the design wind and pressure values for 200m tall building having a uniform square cross section of 30 m * 30 m. 


```python
height_total = 200
storey_height = 4 # height of each floor 
air_density = 1.2 # airdensity in kg/m3
height = np.arange(0, height_total, storey_height)
```

###### Let us compute for terrain category I

```python
a_mean  = 1.18
alpha_mean = 0.18
a_gust = 1.61
alpha_gust = 0.095

umean_1 = a_mean * mean_windspeed * (height/10)**alpha_mean
ugust_1 = a_gust * mean_windspeed * (height/10)**alpha_gust
```

###### Let us plot

```python
plt.figure(num=1, figsize=(8, 6))
plt.plot(umean_1, height, label = 'Mean velocity profile')
plt.plot(ugust_1, height, label = 'Gust velocity profile')
plt.ylabel('Height')
plt.xlabel('wind speed m/s')
plt.title('Wind Speed along height - Terrain category I')
plt.legend()
plt.grid(True)
```

###### Let us compute for terrain category IV

```python
a_mean  = 0.56
alpha_mean = 0.3
a_gust = 1.05
alpha_gust = 0.2

umean_1 = a_mean * mean_windspeed * (height/10)**alpha_mean
ugust_1 = a_gust * mean_windspeed * (height/10)**alpha_gust
```

###### Let us plot

```python
plt.figure(num=2, figsize=(8, 6))
plt.plot(umean_1, height, label = 'Mean velocity profile')
plt.plot(ugust_1, height, label = 'Gust velocity profile')
plt.ylabel('Height')
plt.xlabel('Wind speed m/s')
plt.title('Wind Speed along height - Terrain category IV')
plt.legend()
plt.grid(True)

plt.show()
```
